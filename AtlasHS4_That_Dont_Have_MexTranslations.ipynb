```python
import pandas
```

```python
HSdf = pandas.read_csv("http://raw.githubusercontent.com/cid-harvard/classifications/master/product/HS/Atlas/out/hs92_atlas.csv")
```

```python
HSdf
```

```python
HSdf = HSdf.filter(items=["code","name","level","parent_id"])
```

```python
AtlasHS = HSdf[HSdf["level"] == "4digit"]
```

```python
AtlasHS
```

```python
MexHS = pandas.read_csv("https://raw.githubusercontent.com/cid-harvard/classifications/master/product/HS/Mexico_Prospedia/out/products_mexico_prospedia.csv")
```

```python
MexHS = MexHS[MexHS["level"]=="4digit"]
```

```python
MexHS
```

```python
AtlasHSWithMexMerge = AtlasHS.merge(MexHS,how='left',on="code",suffixes=('_colatlas','_mex'))
```

```python
AtlasHSWithMexMerge
```

```python
AtlasHS_WithNoMexican = AtlasHSWithMexMerge[AtlasHSWithMexMerge["name_mex"].isnull()]
```
